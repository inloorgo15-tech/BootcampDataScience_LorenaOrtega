```python
# ESTRUCTURAS DE CONTROL
registro_ventas = [
    {
        "producto": "Samsung A10",
        "categoria": "Celulares",
        "precio": 3500000,
        "cantidad": 2,
        "total": 7000000
    },
    {
        "producto": "PlayStation 5",
        "categoria": "Videojuegos",
        "precio": 2800000,
        "cantidad": 1,
        "total": 2800000
    },
    {
        "producto": "Balón Nike",
        "categoria": "Deportes",
        "precio": 80000,
        "cantidad": 1,
        "total": 80000
    }
]
for venta in registro_ventas:
    total_original  = venta["total"]
    if total_original > 100000:
        total_con_descuento = total_original * 0.90
    else:
        total_con_descuento = total_original

    print("Producto:", venta["producto"])
    print("Total original: $", total_original)
    print("Total con descuento: $", total_con_descuento)
    print("--------------------------")
```

```python
# FUNCIONES Y MANEJO DE ERRORES
def calcular_ingreso(precio, cantidad):
    try:
        ingreso = precio * cantidad
        return ingreso
    except:
        print("Error: Los valores deben ser numéricos")
        return 0

# Prueba con valores correctos
resultado1 = calcular_ingreso(50000, 3)
print("Ingreso:", resultado1)

# Prueba con un valor incorrecto
resultado2 = calcular_ingreso("cincuenta mil", 3)
print("Ingreso:", resultado2)
```
